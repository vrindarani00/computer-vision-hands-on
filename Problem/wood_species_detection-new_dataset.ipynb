```python
import os
import numpy as np
from PIL import Image

def load_images_from_folders(image_folders):
    images = []
    labels = []
    
    for folder_path in image_folders:
        label = os.path.basename(folder_path)
        
        image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
        
        for filename in image_files:
            image_path = os.path.join(folder_path, filename)
            
            try:
                with Image.open(image_path) as img:
                    image_array = np.array(img)
                    images.append(image_array)
                    labels.append(label)
            except (OSError, IOError):
                print(f"Error reading file: {image_path}")
    
    return images, labels

image_folders = [
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\1",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\2",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\3",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\4",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\5",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\6",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\7",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\8",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\9",
    r"C:\Users\G\DUK\CV_PROJ\github\computer-vision-hands-on\woodentify-master\imgs\images\10"
]


images, labels = load_images_from_folders(image_folders)

print("Number of images:", len(images))
print("Number of labels:", len(labels))

for image, label in zip(images, labels):
    print("Image shape:", image.shape)
    print("Label:", label)
    print()
```

```python
sift = cv2.SIFT_create(nfeatures = 100, contrastThreshold=0.01, edgeThreshold=10)
des_list = []
i = 0
for img in images:
    image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    kp, des = sift.detectAndCompute(image, None)
    des_list.append(des)
    print(i)
    i=i+1
```

```python
stacked = np.vstack(des_list)
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++')
kmeans.fit(stacked)
```

```python
hist_list = []
m = 0
for img in des_list:
    preds = [kmeans.predict(img)]
    hist = np.histogram(preds,bins=5)[0]
    hist_list.append(hist)
    print(m)
    m=m+1
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = tts(hist_list,labels, test_size = 0.25, random_state = 7)
```

```python
model = SVC(kernel = 'rbf')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```
